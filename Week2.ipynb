# JAWABAN

1. cv2.imshow() menampilkan gambar dalam jendela terpisah dengan antarmuka grafis OpenCV dan menggunakan format warna BGR, sementara plt.imshow() menampilkan gambar dalam plot Matplotlib yang terintegrasi dengan notebook atau jendela plot lainnya dan menggunakan format warna RGB. Oleh karena itu, jika menggunakan plt.imshow(), kita perlu mengonversi gambar dari BGR ke RGB agar warna ditampilkan dengan benar.

2. Perbedaan antara tipe data int16 dan int32 dalam pembuatan gambar hitam terletak pada rentang nilai dan penggunaan memori. int16 menggunakan 16 bit, memungkinkan rentang nilai dari -32,768 hingga 32,767, sementara int32 menggunakan 32 bit dengan rentang yang jauh lebih luas. Untuk gambar hitam, perbedaan ini tidak memengaruhi tampilan gambar, tetapi int32 memakan lebih banyak memori dibandingkan int16.

3. google.colab.patches import cv2_imshow digunakan untuk menampilkan gambar yang diproses dengan OpenCV di Google Colab, menggantikan fungsi cv2.imshow() yang tidak dapat digunakan di lingkungan notebook Colab. Fungsi ini memungkinkan gambar ditampilkan langsung dalam sel notebook tanpa perlu mengonversi gambar ke format lain.

4. from skimage import io digunakan untuk membaca dan menampilkan gambar dalam format yang dapat diproses oleh scikit-image, memungkinkan manipulasi dan analisis gambar dalam bentuk array NumPy.

```python
# Membuat gambar kosong
img = np.zeros((400, 400, 3), dtype=np.uint8)
# Mengatur jumlah kotak
num_boxes = 5
# Menggambar kotak dengan ukuran dan warna acak
for _ in range(num_boxes):
    # Menghasilkan koordinat acak untuk sudut kiri atas dan sudut kanan bawah
    x1 = np.random.randint(0, 300)
    y1 = np.random.randint(0, 300)
    x2 = np.random.randint(x1 + 10, 400)
    y2 = np.random.randint(y1 + 10, 400)
    # Menghasilkan warna acak
    color = (np.random.randint(0, 256), np.random.randint(0, 256), np.random.randint(0, 256))
    # Menggambar kotak
    cv.rectangle(img, (x1, y1), (x2, y2), color, -1)  # -1 untuk mengisi kotak
# Menampilkan gambar
plt.figure(figsize=(8, 8))
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.show()
```

```python
# Membaca citra
img = cv.imread('/content/drive/MyDrive/PCVK/kitten.jpeg')

# Memisahkan channel B
b_channel = img[:, :, 0]

# Menentukan posisi dan ukuran rectangle
height, width = b_channel.shape
start_point = (0, height - 50)
end_point = (50, height)
color = 255
thickness = -5

# Menggambar rectangle pada channel B
b_channel_with_rectangle = b_channel.copy()
cv.rectangle(b_channel_with_rectangle, start_point, end_point, color, thickness)

# Menggabungkan kembali channel B ke dalam citra RGB
img_with_rectangle = img.copy()
img_with_rectangle[:, :, 0] = b_channel_with_rectangle

# Menambahkan teks pada gambar
text = 'kitten.jpg'
font = cv.FONT_HERSHEY_SIMPLEX
font_scale = 1
font_color = (0, 255, 0)
thickness = 3
text_size, _ = cv.getTextSize(text, font, font_scale, thickness)
text_width, text_height = text_size

# Posisi teks di kanan bawah
position = (width - text_width - 10, height - 20)
cv.putText(img_with_rectangle, text, position, font, font_scale, font_color, thickness)

# Mengubah citra dari BGR ke RGB
img_with_rectangle_rgb = cv.cvtColor(img_with_rectangle, cv.COLOR_BGR2RGB)

# Menampilkan citra menggunakan matplotlib dengan ukuran lebih kecil
plt.figure(figsize=(6, 4))  # Mengatur ukuran gambar
plt.imshow(img_with_rectangle_rgb)
plt.show()
```

```python
import requests
from io import BytesIO
from PIL import Image
import cv2 as cv
import matplotlib.pyplot as plt

# Unduh gambar dari URL
url_ktp = "https://about.lovia.id/wp-content/uploads/2020/05/150067.jpg"
response = requests.get(url_ktp)
image = np.array(bytearray(response.content), dtype=np.uint8)
image = cv.imdecode(image, cv.IMREAD_COLOR)

# Konversi gambar dari BGR ke RGB
image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Buat figure dan axis
fig, ax = plt.subplots(figsize=(10, 8))
ax.imshow(image_rgb)

# Koordinat dan ukuran rectangle pertama
rect1_start_point = (120, 150)
rect1_width = 90
rect1_height = 14
rect1_color = 'red'
rect1_linewidth = 2

# Tambahkan rectangle pertama ke plot
rect1 = patches.Rectangle(
    rect1_start_point,
    rect1_width,
    rect1_height,
    linewidth=rect1_linewidth,
    edgecolor=rect1_color,
    facecolor='none'
)
ax.add_patch(rect1)

# Koordinat dan ukuran rectangle kedua
rect2_start_point = (120, 166)
rect2_width = 90
rect2_height = 14
rect2_color = 'blue'
rect2_linewidth = 2

# Tambahkan rectangle kedua ke plot
rect2 = patches.Rectangle(
    rect2_start_point,
    rect2_width,
    rect2_height,
    linewidth=rect2_linewidth,
    edgecolor=rect2_color,
    facecolor='none'
)
ax.add_patch(rect2)

# Tampilkan gambar
plt.show()
```
